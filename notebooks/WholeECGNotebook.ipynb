```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("../datasets/ECG200_TRAIN.txt", "r")
```

```python
X = []
Y = []
for line in file.readlines():
    Y.append([float(i) for i in line.split()][0])
    X.append([float(i) for i in line.split()][1:])

X = np.array(X, dtype='float32')
Y = np.array(Y, dtype='int8')
```

```python
print(Y[0])
print(X[0])
```

```python
index = 5
print(Y[index])
plt.plot([i for i in range(len(X[index]))], X[index])
plt.show()
```
